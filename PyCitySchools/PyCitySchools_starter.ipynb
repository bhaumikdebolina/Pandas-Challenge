```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Calculate the total number of schools
total_schools=school_data["school_name"].count()
total_schools

#Calculate the total number of students
total_students=student_data["student_name"].count()
total_students

#total budget
total_budget=school_data["budget"].sum()
total_budget

#avg math score
avg_math_score=student_data["math_score"].mean()
avg_math_score

#avg reading
avg_reading_score=student_data["reading_score"].mean()
avg_reading_score

#passing math student
percent_math_pass=(school_data_complete["math_score"]>70).sum() / school_data_complete["math_score"].count()*100
percent_math_pass

#passing reading
percent_reading_score=(school_data_complete["reading_score"]>70).sum() / school_data_complete["reading_score"].count()*100
percent_reading_score

#percent overall pass in both subjects
overall_pass=((avg_math_score + avg_reading_score) /2)
overall_pass

#dataframe
District_Summary = pd.DataFrame({"Total School":[total_schools], "Total Student":[total_students],"Total Budget":[total_budget], "Average Math Score":[avg_math_score], "Average Reading Score":[avg_reading_score],"% Passing Math":[percent_math_pass],"%Passing Reading":[percent_reading_score],"%Overall Passing Rate":[overall_pass]})
District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type

  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score

* Create a dataframe to hold the above results

```python
#school name 
school_count = len(school_data_complete['school_name'].unique())
school_count
```

```python
#school name and type
school_types=school_data.set_index(["school_name"])["type"]
#school_types.head()
```

```python
#Total Students
student_per_school=school_data_complete["school_name"].value_counts()
student_per_school.head()
```

```python
#total budget
school_budget=school_data_complete.groupby(["school_name"])["budget"].mean()
school_budget.head()
```

```python
#per student budget
Per_Student_Budget=school_budget/student_per_school
Per_Student_Budget
```

```python
#Average Math Scorecper School
avgmath_perschl=school_data_complete.groupby(["school_name"])["math_score"].mean()
avgmath_perschl
```

```python
#average reading score
avgreading_perschl=school_data_complete.groupby(["school_name"])["reading_score"].mean()
avgreading_perschl
```

```python
#Passing math per school
passing_math_score=school_data_complete.loc[school_data_complete["math_score"]>=70]
passing_math_score

group_math_schl=passing_math_score["school_name"].value_counts()
group_math_schl.head()

percent_math=group_math_schl/student_per_school*100
percent_math
```

```python
#passing read per school
passing_read_score=school_data_complete.loc[school_data_complete["reading_score"]>=70]
passing_read_score

group_read_schl=passing_read_score["school_name"].value_counts()
group_read_schl

percent_reading=group_read_schl/student_per_school*100
percent_reading
```

```python
#overall passing
overall_pass = (percent_reading + percent_math)/student_per_school
overall_pass
```

```python
#school summary dataframe
School_summary_df = pd.DataFrame({"School Type":school_types, "Total Student":student_per_school, "Total School Budget":school_budget, "Per Student Budget":Per_Student_Budget, "Average Math Score":avgmath_perschl, "Average Reading Score":avgreading_perschl, "% Passing Math":percent_math, "% Passing Reading":percent_reading, "% Overall Passing":overall_pass})
School_summary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
School_summary_df.head()
```

```python
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

scores_by_spending = School_summary_df[["School Type","Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading","% Overall Passing"]]
scores_by_spending["Spending Score"] = pd.cut(School_summary_df["Per Student Budget"], bins, labels=group_names)
scores_by_spending = scores_by_spending.groupby(["Spending Score"])
scores_by_spending.mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
population_size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_by_size = School_summary_df[["School Type","Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading","% Overall Passing"]]
school_by_size["School size"] = pd.cut(School_summary_df["Total Student"], population_size_bins, labels=group_names)
school_by_size = school_by_size.groupby(["School size"])
school_by_size.mean()
```
